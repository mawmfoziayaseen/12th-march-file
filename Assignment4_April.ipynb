# Convert SampleData.xlsx file into a csv file and save it in a CSV named folder in the current  directory.  

```python
import pandas as pd
import os

# Load the Excel file
df = pd.read_excel("SampleData.xlsx")

# Create a directory named 'CSV' if it doesn't exist
csv_folder = "CSV"
if not os.path.exists(csv_folder):
    os.makedirs(csv_folder)

# Save DataFrame to a CSV file in the 'CSV' folder
df.to_csv(os.path.join(csv_folder, "SampleData.csv"), index=False)
```
