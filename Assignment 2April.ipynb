#Task 4: Mark fail to those students who got less than 50 marks 

```python
df['Result'] = df['Total'].apply(lambda x: 'Pass' if x >= 50 else 'Fail')
```

```python
#: Make a new column as GRADE and assign grade according to this manner 
```

```python
df.loc[(df['Total\n(100)'] >= 90) & (df['Total\n(100)'] <= 100), 'GRADE'] = 'A+'
df.loc[(df['Total\n(100)'] >= 80) & (df['Total\n(100)'] < 90), 'GRADE'] = 'A-'
df.loc[(df['Total\n(100)'] >= 70) & (df['Total\n(100)'] < 80), 'GRADE'] = 'B'
df.loc[(df['Total\n(100)'] >= 60) & (df['Total\n(100)'] < 70), 'GRADE'] = 'C'
df.loc[(df['Total\n(100)'] >= 50) & (df['Total\n(100)'] < 60), 'GRADE'] = 'D'
df.loc[df['Total\n(100)'] < 50, 'GRADE'] = 'Fail'

df
```

```python
#: Rank your data frame with respect to total marks. 
```

```python
# Rank the DataFrame based on 'Total' column
df['Rank'] = df['Total'].rank(ascending=False)

# Display the DataFrame with ranks
print(df)
```

```python
# Calculate the average of total marks. 
```

```python
# Calculate the average of total marks
average_total_marks = df['Total'].mean()

# Display the average
print("Average Total Marks:", average_total_marks)
```

```python
# Display the name of the first three position holders.  
```

```python
# Sort the DataFrame by Total marks in descending order
top_students = df.sort_values(by='Total', ascending=False)

# Display the names of the first three position holders
print("Names of the first three position holders:")
print(top_students['Name of Student'].head(3))
```
