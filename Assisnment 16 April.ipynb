```python
#Use subplot to plot 4 Plots of random integer numbers in ranges of:
#1 to 100, 200 to 300, 300 to 400, 400 to 500.
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random integer numbers in specified ranges
data1 = np.random.randint(1, 101, 100)  # Range: 1 to 100
data2 = np.random.randint(200, 301, 100)  # Range: 200 to 300
data3 = np.random.randint(300, 401, 100)  # Range: 300 to 400
data4 = np.random.randint(400, 501, 100)  # Range: 400 to 500

# Create subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Plot data
axs[0, 0].hist(data1, bins=20, color='skyblue', edgecolor='black')
axs[0, 0].set_title('1 to 100')

axs[0, 1].hist(data2, bins=20, color='salmon', edgecolor='black')
axs[0, 1].set_title('200 to 300')

axs[1, 0].hist(data3, bins=20, color='lightgreen', edgecolor='black')
axs[1, 0].set_title('300 to 400')

axs[1, 1].hist(data4, bins=20, color='gold', edgecolor='black')
axs[1, 1].set_title('400 to 500')

plt.tight_layout()
plt.show()
```
