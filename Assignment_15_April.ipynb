# Task1

```python
# Generate 1000 Gaussian random numbers using randn command and plot. 
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate 1000 Gaussian random numbers
random_numbers = np.random.randn(1000)

# Create the plot
plt.hist(random_numbers, bins=100, density=True)  # Adjust bins for better visualization
plt.xlabel('Random Numbers')
plt.ylabel('Probability Density')
plt.title('Distribution of 1000 Gaussian Random Numbers')
plt.grid(True)

# Show the plot
plt.show()
```

# Task2

```python
#Generate 1000 Random integers and plot a line chart using these values. Change the colour of the line to red.
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate 1000 random integers between 0 and 1000 (inclusive)
random_integers = np.random.randint(0, 1001, size=1000)

# Create the line plot with red color
plt.plot(random_integers, color='red')

# Set labels and title
plt.xlabel('Index')
plt.ylabel('Random Integer')
plt.title('Line Plot of 1000 Random Integers')

# Show the plot
plt.show()
```

# Task3

```python
#Plot sine wave with frequency 50.
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define sampling rate and duration
sampling_rate = 1000  # Samples per second
duration = 1  # Seconds

# Calculate time vector
t = np.linspace(0, duration, int(sampling_rate * duration))

# Generate sine wave with frequency 50 Hz
frequency = 50  # Hz
y = np.sin(2 * np.pi * frequency * t)  # Radians for sine function

# Plot the sine wave
plt.plot(t, y)

# Set labels and title
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Sine Wave (Frequency: 50 Hz)')

# Show the plot
plt.grid(True)
plt.show()
```

# Task 4

```python
# Plot cosine wave with frequency 25. 
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define sampling rate and duration
sampling_rate = 1000  # Samples per second
duration = 1  # Seconds

# Calculate time vector
t = np.linspace(0, duration, int(sampling_rate * duration))

# Generate cosine wave with frequency 25 Hz
frequency = 25  # Hz
y = np.cos(2 * np.pi * frequency * t)  # Radians for cosine function

# Plot the cosine wave
plt.plot(t, y)

# Set labels and title
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Cosine Wave (Frequency: 25 Hz)')

# Show the plot
plt.grid(True)
plt.show()
```
